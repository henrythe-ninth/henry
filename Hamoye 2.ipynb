```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the dataset

data = pd.read_csv ('c:/datasets/energydata_complete.csv')
```

```python
#displaying the first 5 rows of the dataset

data.head()
```

```python
data.shape
```

```python
#dropping the colunms date and lights and creatinig a new dataset

data_new = data.drop(['date', 'lights'], axis = 1)
```

```python
#importing scaler MinMaxScaler 

from sklearn.preprocessing import MinMaxScaler 
```

```python
#scaling data 

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data_new)
scaled_data = pd.DataFrame(scaled_data, columns = data_new.columns )
```

```python
scaled_data
```

```python
#checking for empty values

scaled_data.isnull().sum()
```

```python
scaled_data.describe()
```

```python
#finding the correlation and plotting the heatmap

corr = scaled_data.corr()
corr

f, ax = plt.subplots(figsize=(50, 10))
sns.heatmap(corr,cmap='coolwarm',linewidths=2.0, annot=True)
```

```python
#filteriing the data

df = scaled_data.drop(columns= ['rv2'])
```

```python
#x is the features while y is the target variable

x = df.drop (columns = ['Appliances'])
y = df.Appliances
```

```python
#splitinng the dataset innto train and test 

from sklearn.model_selection import train_test_split 
xtrain, xtest, ytrain, ytest = train_test_split (x, y, random_state=42, test_size = 0.3)
```

```python
#importing metrics 

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# linear regression model

from sklearn.linear_model import LinearRegression
linear = LinearRegression (normalize = True)

#fitting the model to the training set
linear.fit(xtrain, ytrain)

#predictions
y_pred = linear.predict (xtest)
```

```python
importance = linear.coef_
importance
```

```python
#printing the error values for linear model 

mae = mean_absolute_error(ytest, y_pred)
rmse = np.sqrt(mean_squared_error(ytest, y_pred))
r2 = r2_score(ytest, y_pred)



print ('The MAE is:' , round(mae, 3))
print ('The RMSE is:' ,round(rmse, 3))
print ('The r2 score is:', round(r2, 3))
```

```python
#ridge regression

from sklearn.linear_model import Ridge
ridge = Ridge(alpha = 0.4)
ridge.fit(xtrain, ytrain)

#prediction
y_ridge = ridge.predict(xtest)
```

```python
#printing the error values for linear model 

mae_ridge = mean_absolute_error(ytest, y_ridge)
rmse_ridge = np.sqrt(mean_squared_error(ytest, y_ridge))
r2_ridge = r2_score(ytest, y_ridge)



print ('The MAE is:' , round(mae_ridge, 3))
print ('The RMSE is:' ,round(rmse_ridge, 3))
print ('The r2 score is:', round(r2_ridge, 3))
```
